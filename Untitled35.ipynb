```r
install.packages("nardl")
```

```r
library(nardl)
```

# CARGAR DATOS

```r
D = read.csv("/content/DATOSMACROEUROPA70-18 (3).csv")
E = D[, c("URX", "YER")]
```

```r
LE= log(E)
LEsna =na.omit(LE)
```

# DEFINIR MODELO

```r
MOD.DES= nardl (URX~YER, data=LEsna, ic="aic", maxlag=TRUE, graph=TRUE, case=3)
```

#ANALISIS COINTEGRACION OCULTA

```r
summary(MOD.DES)
```
