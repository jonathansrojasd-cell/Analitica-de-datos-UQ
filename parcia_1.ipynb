```python
# PUNTO 4
# Departamentos
departamentos = {
    "Risaralda":66,
    "Caldas":17,
    "Tolima":73,
    "Quindio":63
}

resultados = {}

for nombre, codigo in departamentos.items():
    
    # Mujeres desempleadas 18-25
    D = df_desocupados[(df_desocupados['DSI']==1) &
                       (df_desocupados['P3271']==2) &
                       (df_desocupados['P6040']>=18) &
                       (df_desocupados['P6040']<=25) &
                       (df_desocupados['DPTO']==codigo)]['FEX_C18'].sum()
    
    # PEA mujeres 18-25
    PEA = df_ft[(df_ft['FT']==1) &
                (df_ft['P3271']==2) &
                (df_ft['P6040']>=18) &
                (df_ft['P6040']<=25) &
                (df_ft['DPTO']==codigo)]['FEX_C18'].sum()
    
    TD = D/PEA
    resultados[nombre] = TD
    
    print(f"Tasa de desempleo mujeres 18-25 en {nombre}: {TD*100}%")

# Departamento con menor tasa
menor = min(resultados, key=resultados.get)

print("------------------------------------------------")
print(f"El departamento con menor tasa de desempleo de mujeres entre 18 y 25 años es: {menor}")
print(f"Tasa: {resultados[menor]*100}%")
```
